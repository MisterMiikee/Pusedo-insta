## Statistics per Numerical and Categorical Variable

### Numerical Descriptive Stats & Histograms

```python
# Filter numeric columns
numeric_cols = df_copy.select_dtypes(include=['number']).columns
```

```python
df_copy[numeric_cols].describe()
```

```python
fig = plt.figure(figsize=(12, 12))

for i, col in enumerate(numeric_cols):
    ax = fig.add_subplot(3, 3, i+1)
    sns.histplot(df_copy[col], ax=ax, kde=True)
    ax.set_title(col, fontsize=15, y=1.02)
    ax.set_xlabel('')

plt.suptitle('Histogram per Numeric Category', fontsize=20, fontweight='bold')

sns.despine()
plt.tight_layout()
plt.savefig('/content/numeric_plots.png')
plt.show()
```

### Categorical Bar Plots

```python
# Filter categorical columns
categorical_cols = df_copy.select_dtypes(include=['object']).columns
```

```python
fig = plt.figure(figsize=(20, 20))
for i, col in enumerate(categorical_cols):
    # Get value counts and select top 10 categories
    top_categories = df_copy[col].value_counts().nlargest(10)

    ax = fig.add_subplot(5, 5, i+1)
    sns.barplot(x=top_categories.values, y=top_categories.index, ax=ax)
    ax.set_title(col, fontsize=15, y=1.02)
    ax.set_xlabel('Count')
    ax.set_ylabel('')

plt.suptitle('Distribution per Categorical Category', fontsize=20, fontweight='bold', y=1)

sns.despine()
plt.tight_layout()
plt.savefig('categorical_plots.png')
plt.show()
```
